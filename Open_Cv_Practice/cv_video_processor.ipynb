```python
#capture the video through the webcam
video =  cv2.VideoCapture(0)

a=1
while True:
    a = a+1
    #now add the wiondow for seeing the video
    check ,frame = video.read()
    print(frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Capturing the video",gray)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
print(a)
video.release()
cv2.destroyAllWindows()
```
